```python
import whisper

# Load the Whisper model
model = whisper.load_model("base")

# Transcribe the audio
result = model.transcribe(audio_path)

# Print the transcription with timestamps
for segment in result["segments"]:
    print(f"Start: {segment['start']}, End: {segment['end']}, Text: {segment['text']}")
```

```python
def semantic_chunking(segments, max_chunk_duration=10):
    """
    Group segments into chunks based on duration or sentence boundaries.
    """
    chunks = []
    current_chunk = []
    current_duration = 0

    for segment in segments:
        if current_duration + (segment['end'] - segment['start']) <= max_chunk_duration:
            current_chunk.append(segment)
            current_duration += segment['end'] - segment['start']
        else:
            chunks.append(current_chunk)
            current_chunk = [segment]
            current_duration = segment['end'] - segment['start']

    if current_chunk:
        chunks.append(current_chunk)

    return chunks

# Apply semantic chunking
chunks = semantic_chunking(result["segments"])
for i, chunk in enumerate(chunks):
    print(f"Chunk {i+1}:")
    for segment in chunk:
        print(f"  Start: {segment['start']}, End: {segment['end']}, Text: {segment['text']}")
```

```python
from pydub import AudioSegment

# Load the audio file
audio = AudioSegment.from_file(audio_path)

# Split the audio into chunks
for i, chunk in enumerate(chunks):
    start_time = chunk[0]['start'] * 1000  # Convert to milliseconds
    end_time = chunk[-1]['end'] * 1000
    chunk_audio = audio[start_time:end_time]
    chunk_audio.export(f"chunk_{i+1}.wav", format="wav")
    print(f"Exported chunk {i+1} (start: {start_time/1000}s, end: {end_time/1000}s)")
```

```python
import json

# Prepare the output data
output = []
for i, chunk in enumerate(chunks):
    chunk_data = {
        "chunk_id": i + 1,
        "start_time": chunk[0]['start'],
        "end_time": chunk[-1]['end'],
        "text": " ".join([segment['text'] for segment in chunk]),
        "audio_file": f"chunk_{i+1}.wav"
    }
    output.append(chunk_data)

# Save the output to a JSON file
with open("output.json", "w") as f:
    json.dump(output, f, indent=4)
print("Output saved to output.json")
```

```python
import cv2
import numpy as np
import librosa
from sklearn.cluster import KMeans
from face_recognition import face_locations
from collections import defaultdict

def extract_semantic_segments(video_path):
    cap = cv2.VideoCapture(video_path)
    fps = cap.get(cv2.CAP_PROP_FPS)
    
    # Scene Detection
    scene_transitions = []
    prev_frame = None
    frame_count = 0

    # Action Recognition (Optical Flow)
    action_segments = []
    prvs = None

    # Face Clustering
    face_embeddings = []
    face_timestamps = []

    # Audio Analysis
    audio, sr = librosa.load(video_path, sr=None)
    mfcc = librosa.feature.mfcc(y=audio, sr=sr)
    audio_changes = np.mean(np.abs(np.diff(mfcc, axis=1)), axis=0)

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Scene Change Detection
        if frame_count % int(fps) == 0:  # Check every second
            gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            if prev_frame is not None:
                hist_diff = cv2.compareHist(
                    cv2.calcHist([gray], [0], None, [256], [0,256]),
                    cv2.calcHist([prev_frame], [0], None, [256], [0,256]),
                    cv2.HISTCMP_BHATTACHARYYA
                )
                if hist_diff > 0.8:  # Threshold
                    scene_transitions.append(frame_count/fps)
            prev_frame = gray

        # Action Recognition (Optical Flow)
        if frame_count % 2 == 0:  # Process every 2nd frame
            gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            if prvs is not None:
                flow = cv2.calcOpticalFlowFarneback(prvs, gray, None, 0.5, 3, 15, 3, 5, 1.2, 0)
                magnitude = np.sqrt(flow[...,0]**2 + flow[...,1]**2)
                if np.mean(magnitude) > 5.0:  # Motion threshold
                    action_segments.append(frame_count/fps)
            prvs = gray

        # Face Detection
        if frame_count % int(fps) == 0:  # Check every second
            face_locs = face_locations(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
            if face_locs:
                face_timestamps.append(frame_count/fps)

        frame_count += 1

    cap.release()

    # Process Results
    semantic_segments = {
        "scenes": _cluster_time_points(scene_transitions, fps),
        "action": _cluster_time_points(action_segments, fps),
        "faces": _cluster_time_points(face_timestamps, fps),
        "audio": _find_audio_transitions(audio_changes, sr, fps)
    }

    return semantic_segments

def _cluster_time_points(points, fps, threshold=2):
    if not points:
        return []
    points = np.array(points).reshape(-1,1)
    kmeans = KMeans(n_clusters=min(5, len(points))).fit(points)
    clusters = defaultdict(list)
    for idx, label in enumerate(kmeans.labels_):
        clusters[label].append(points[idx][0])
    return [max(cluster) for cluster in clusters.values()]

def _find_audio_transitions(audio_changes, sr, fps):
    peaks = librosa.util.peak_pick(audio_changes, pre_max=3, post_max=3, pre_avg=3, post_avg=5, delta=0.5, wait=10)
    return [(peak * 512 / sr) for peak in peaks]
```
